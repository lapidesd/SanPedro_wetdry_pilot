```python
import pandas as pd
import numpy as np
import dataretrieval.nwis as nwis
import matplotlib.pyplot as plt
```

```python
ramseyflow = nwis.get_record(sites='09470750', service='iv',start='2005-07-16',end='2024-01-01',
                             parameterCd='00060').reset_index()
ramseyflow['datetime'] = pd.to_datetime(ramseyflow['datetime'])
ramseyflow.set_index('datetime',inplace=True)
ramseyflow['m3/d'] = ramseyflow['00060_Mean']*0.028316831998814504*60*60*24
ramseyflow['mm/d'] = ramseyflow['m3/d']/1.0645e+7*1000
```

```python
ramseyhydro = pd.read_csv('../data/Ramsey/RamseyHydroData.csv')
ramseyhydro['Unnamed: 0'] = pd.to_datetime(ramseyhydro['Unnamed: 0'])
ramseyhydro.set_index('Unnamed: 0',inplace=True)
```

```python
ramseyflow['mm/d'].plot()
plt.yscale('log')
ramseyhydro['Q [mm/d]'].plot()
plt.yscale('log')
```
