```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
def import_data(csv):
    # function than imports csv data to dataframe and 
    # converts datetime data to a datetime type

    df = pd.read_csv(csv)
    datecol = df.columns[0]
    df[datecol] = pd.to_datetime(df[datecol])
    df.set_index(datecol,inplace=True)
    return df
```

```python
# import data
Ramsey = import_data('../data/Ramsey/RamseyHydroData.csv')
Cienega = import_data('../data/Cienega/CienegaHydroData.csv')
```

```python
year = 2010
fig,ax = plt.subplots(1,figsize = (8,4))
Ramsey.loc['1-1-'+str(year):'1-1-'+str(year+1)]['Q [mm/d]'].plot(ax=ax)
ax.set_xlabel('')
ax.set_yscale('log')
ax.set_ylabel('Streamflow [mm/d]')
ax.set_title('Ramsey Canyon')
```
