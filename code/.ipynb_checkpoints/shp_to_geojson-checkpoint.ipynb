```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import math
import numpy as np
import os
```

```python
# import survey dates
cienega_dates = pd.read_csv('../data/Cienega/Cienega_survey_dates.csv')
# fill nan values with survey dates in left column
cienega_dates['Cienega date'] = np.where(cienega_dates['Cienega date'].isna(),
                                        cienega_dates['mmddyyyy'],
                                        cienega_dates['Cienega date'])
cienega_dates
```

```python
dataloc = '../data/Cienega/wetdry/' # location where raw data are stored
monthdict = {'Dec':12,'Jan':1,'Feb':2,'Mar':3,
            'Apr':4,'May':5,'Jun':6,'Jul':7,
            'Aug':8,'Sep':9,'Oct':10,'Nov':11,}

# loop through data to convert to geojson files named by survey date
for y in [yy for yy in os.listdir(dataloc) if yy.startswith('2')]:
    for date in [d for d in os.listdir(dataloc+y) if d!='.DS_Store']:
        print(y,date)
        try:
        
            fnames = os.listdir(dataloc+y+'/'+date)
            fname = [f for f in fnames if (f.startswith('ccflow')) and ('all' not in f)]
            surveydata = gpd.read_file(dataloc+y+'/'+date+'/'+fname[0].split('.')[0]+'.shp')
        except:
            try:
                lname = [c for c in os.listdir(dataloc+y+'/'+date) if c!='.DS_Store'][0]
                lname = lname.split('.')[0]
                fnames = os.listdir(dataloc+y+'/'+date+'/'+lname+'/')
                fname = [f for f in fnames if (f.startswith('ccflow')) and ('all' not in f)]
                surveydata = gpd.read_file(dataloc+y+'/'+date+'/'+lname+'/'+fname[0].split('.')[0]+'.shp')
            except:
                lname = [c for c in os.listdir(dataloc+y+'/'+date) if c!='.DS_Store']
                lname = [l for l in lname if (l.startswith('ccflow')) & ('all' not in l) & ('dam' not in l)][0]
                lname = lname.split('.')[0]
                fnames = os.listdir(dataloc+y+'/'+date+'/'+lname+'/')
                fname = [f for f in fnames if (f.startswith('ccflow')) and ('all' not in f)]
                surveydata = gpd.read_file(dataloc+y+'/'+date+'/'+lname+'/'+fname[0].split('.')[0]+'.shp')
                
        # get date of present survey
        mval = monthdict[date.split(' ')[0]] # numerical value associated with given month
        surveydate = cienega_dates[(cienega_dates['Cienega date'].str.contains('/'+y)) & # survey dates for given year only
                      ((cienega_dates['Cienega date'].str.startswith(str(mval))) | # survey dates for given month only
                       (cienega_dates['mmddyyyy'].str.startswith(str(mval))))
                       ]['Cienega date'].values[0]
        surveydate = surveydate.replace('/','-')
        surveydata.to_file(dataloc+'survey_data/'+surveydate+'.geojson', driver="GeoJSON")  
```
